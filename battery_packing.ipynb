```python
x = df.drop(columns=['Voltage(V)'], axis=1)
y = df['Voltage(V)']
```

```python
df.dtypes
```

```python
# Convert 'h/m/s' column to numerical (total seconds)
def convert_hms_to_seconds(hms_str):
    if isinstance(hms_str, str):
        h, m, s = map(int, hms_str.split(':'))
        return h * 3600 + m * 60 + s
    return hms_str

x['h/m/s'] = x['h/m/s'].apply(convert_hms_to_seconds)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score = r2_score(y_test,y_pred)
```
