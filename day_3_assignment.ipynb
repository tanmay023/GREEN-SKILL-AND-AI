```python
import pandas as pd
music_df = pd.read_csv('music.csv')
print(music_df)
```

```python
total_no_of_plays_for_each_country = music_df.groupby('country')['plays'].sum()
print(total_no_of_plays_for_each_country)
```

```python
total_no_of_plays_for_each_country_and_genre = music_df.groupby(['country','genre'])['plays'].sum()
print(total_no_of_plays_for_each_country_and_genre)
```

```python
date = pd.date_range(start='2021-10-01',end='2021-10-10')

data = {
    'Date':date,
    'Measurement':[16.0,13.0,14.0,12.0,None,None,None,8.0,7.0,5.0]
}

df = pd.DataFrame(data)
print(df)

mean = df['Measurement'].mean()

df.fillna(method='ffill',limit=2,inplace=True) # FILLING FIRST @ CONSECATIVE VALUES WITH FORWARD FILL
df.fillna(mean,inplace=True) # FILLING LAST @ AND REMAINING ONE WITH MEAN/AVG
print(df)
```

```python
date = pd.date_range(start='2021-10-01',end='2021-10-10')

data = {
    'Date':date,
    'Measurement':[16.0,13.0,14.0,12.0,None,None,None,8.0,7.0,5.0]
}

df = pd.DataFrame(data)
print(df)

# OR WITH THE FUNCTION

def fill_missing_values():
  try:
    df.fillna(method="ffill",limit=2,inplace=True)
    df.fillna(df['Measurement'].mean(),inplace=True)
  except:
    pass
```

```python
fill_missing_values()
print(df)
```
