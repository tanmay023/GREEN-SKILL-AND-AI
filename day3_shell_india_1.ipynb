```python
import pandas as pd
import numpy as np

data = {
    "Energy Source":["Solar","Wind","Hydropower","Geothermal","Biomass","Nuclear"],
    "Energy Consumption":[1200,np.nan,2000,np.nan,2500,3200],
    "Cost (Million $)":[200,400,np.nan,150,250,np.nan]
}

energy_df = pd.DataFrame(data)
print("Original data with missing values:")
print(energy_df)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
energy_df['Energy Consumption'].fillna(energy_df['Energy Consumption'].mean(),inplace=True)
energy_df['Cost (Million $)'].fillna(energy_df['Cost (Million $)'].mean(),inplace=True)
print("Data after filling misssing values:")
print(energy_df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Energy Source":["Solar","Wind","Hydropower","Geothermal","Biomass","Nuclear"],
    "Energy Consumption":[1200,np.nan,2000,np.nan,2500,3200],
    "Cost (Million $)":[200,400,np.nan,150,250,np.nan]
}

energy_df = pd.DataFrame(data)
print("Original data with missing values:")
print(energy_df)
```

```python
forward_fill_df = energy_df.fillna(method='ffill')
print("Data after forward fill:")
print(forward_fill_df)
```

```python
# NORMALIZATION (MIN-MAX SCALING)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
energy_df[["Energy Consumption", "Cost (Million $)"]] = scaler.fit_transform(energy_df[["Energy Consumption", "Cost (Million $)"]])
print("Normalized data:")
print(energy_df)
```

```python
# SATNDARIZATION Z-SCALING

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
energy_df[['Energy Consumption', 'Cost (Million $)']] = scaler.fit_transform(energy_df[['Energy Consumption', 'Cost (Million $)']])

print("DATA AFTER STANDARDIZATION(Z-SCALING):")
print(energy_df)
```
