```python
from sklearn.preprocessing import StandardScaler
scaled_data = StandardScaler().fit_transform(df)
```

```python
print(pd.DataFrame(scaled_data,columns=df.columns).head())
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method='ward'))
plt.axhline(y=80, color='r', linestyle='--')
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.xlabel('Customers')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=4,linkage='ward')
hierarchical_cluster.fit_predict(scaled_data)
print(df.head())
```

```python
print(df)
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.scatterplot(x='co2_emissions',y='waste_production',hue='cluster',data=df,palette='viridis',s=100,edgecolor='k')
plt.title('Hierarchical Clustering',fontsize=15)
plt.xlabel('co2_emissions',fontsize=10)
plt.ylabel('waste_production',fontsize=10)
plt.legend(fontsize=12)
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(scaled_data, hierarchical_cluster.labels_)
print("Silhouette Score:", sil_score)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create box plots for each numerical feature against the cluster
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

numerical_cols = ['co2_emissions', 'waste_production', 'gdp', 'population']

for i, col in enumerate(numerical_cols):
    sns.boxplot(x='cluster', y=col, data=df, palette='viridis', ax=axes[i], hue='cluster', legend=False)
    axes[i].set_title(f'Distribution of {col} across Clusters')
    axes[i].set_xlabel('Cluster')
    axes[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```
