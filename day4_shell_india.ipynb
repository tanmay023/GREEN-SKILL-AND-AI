```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from matplotlib.lines import lineStyles
months = ['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption = [1200,1300,1100,1500,1400,1600]

plt.plot(months,energy_consumption,marker= 'D',color='g',linestyle='--',ms=10,mec='k',mfc='m') # used to draw points (MARKERS) draws line from point to point (x-label,y-label)
plt.title("Energy Consumption")
plt.xlabel("Months")
plt.ylabel("Energy Consumption (MWh)")
plt.show()
```

```python
energy_sources = ['Solar','Wind','Hydropower','Biomass']
energy_value = [1200,1400,2900,2500]

plt.bar(energy_sources,energy_value,color='green')
plt.title("Energy Consumption")
plt.xlabel("Energy Sources")
plt.ylabel("Energy Consumption (MWh)")
plt.grid()
plt.show()
```

```python
plt.barh(energy_sources,energy_value,color='green')
plt.title("Energy Consumption")
plt.xlabel("Energy Sources")
plt.ylabel("Energy Consumption (MWh)")
plt.grid()
plt.show()
```

```python
from IPython.core.magics import auto
plt.pie(energy_value,labels=energy_sources,autopct='%1.1f%%',startangle=90)

plt.title('Energy Consumption by source')
plt.show()
```

```python
carbon_emission = [200,500,450,300]
plt.scatter(energy_value,carbon_emission,color='b')

plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emission (kg co2)')
plt.show()
```

```python
plt.bar(energy_sources,energy_value,edgecolor='b',color='g')
plt.title("Customisez energy consumption by source",fontsize=15,fontweight='bold')
plt.ylabel("Energy Consumption (MWh)",fontsize=12)
plt.xlabel("Energy Sources",fontsize=12)

plt.grid(True)
plt.legend(['energy_consumption'],loc='upper left')
plt.show()
```
