```python
city_data =[
    {"name": "PUNE",
    "temp": 25,
    "carbon_footprint": 500.75,
    "is_sustainable": False
     },
    {"name": "MUMBAI",
    "temp": 28,
    "carbon_footprint": 650.75,
    "is_sustainable": False
     },
    {"name": "DELHI",
    "temp": 15,
    "carbon_footprint": 700.75,
    "is_sustainable": False
     },
    {"name": "NAGPUR",
    "temp": 30,
    "carbon_footprint": 510.75,
    "is_sustainable": False
     },
    {"name": "YAVATMAL",
    "temp": 27,
    "carbon_footprint": 400.75,
    "is_sustainable": True
     }
]

print(city_data)

high_temp_threshold = 25
high_temp_cities = [city for city in city_data if city["temp"]>high_temp_threshold]
print(high_temp_cities)

for city in city_data:
  print(city)

total_carbon_footprint = 0 # Initialize total_carbon_footprint before loop
for city in city_data:
  total_carbon_footprint += city["carbon_footprint"]

print(f"total carbon footprint: {total_carbon_footprint}")
avg_carbonfootprint = total_carbon_footprint/len(city_data)
print(f"average carbon footprint: {avg_carbonfootprint:.2f}KG CO2")

sustainable_threshold = 400
sustainable_cities = list(filter(lambda city: city["carbon_footprint"]< sustainable_threshold, city_data))
for city in sustainable_cities:
  print(city)
```

```python
highest_foot_print_city = max(city_data, key=lambda city: city["carbon_footprint"])
print(f"city with highest carbon footprint: {highest_foot_print_city}")
```

```python
import pandas as pd

renewable_sources = ["SOLAR","WIND","HYDROPOWER","GEOTHERMAL","BIOMASS"]

data = {
    "project":["Solar Farm A","Wind Turbine X","Hydropower Y","Solar Roof z","Geothermal Plant P"],
    "Technology":["Solar","Wind","Hydropower","Solar","Geothermal"],
    "Capacity (MW)": [150,300,200,50,100],
    "Cost(Million $)":[200,400,350,100,250],
    "Location":["California","Texus","Washington","Nevada","Idaho"],
    "Completion Year":[2023,2024,2022,2025,2023]
}

```

```python
renewable_series = pd.Series(renewable_sources)
print("Renewable Energy Sources:")
print(renewable_series)
```

```python
projects_df = pd.DataFrame(data)

print("\nGreen Technology Project DataFrame:")
print(projects_df)
```

```python
print("\nList of Project")
print(projects_df["project"])
```

```python
high_capacity_projects = [projects_df["Capacity (MW)"]>100]
print("\nProjects with the capacity greater than 100 MW:")
print(high_capacity_projects)
```

```python
projects_df["Cost per MW"] = projects_df["Cost(Million $)"]/projects_df["Capacity (MW)"]
print("\nDataFrame with Cost per MW:")
print(projects_df)
```

```python
total_capacity = projects_df["Capacity (MW)"].sum()
total_cost = projects_df["Cost(Million $)"].sum()

print(f"\nTotal Capacity of all the projects: {total_capacity} MW")
print(f"\nTotal Cost of all projects: {total_cost} Million $")
```

```python
grouped_data = projects_df.groupby("Technology")["Capacity (MW)"].sum()
print("\nTotal Capacity by Technology:")
print(grouped_data)
```
