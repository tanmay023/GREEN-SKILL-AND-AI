```python
data = pd.DataFrame({'Name':['Alice','Bob','Charlie','David'],'Age':[25,50,35,48],'Salary':[10000,96000,54000,52000]})
print(data)
sorted_data = data.sort_index(axis=0,ascending=False)
print(sorted_data)
```

```python
sorted_data = data.sort_index(axis=1,ascending=False)
print(sorted_data)
```

```python
sorted_data = data.sort_values(by='Age',ascending=True)
print(sorted_data)
```

```python
sorted_data = data.sort_values(by=['Age','Salary'],ascending=[True,False])
print(sorted_data)
```
